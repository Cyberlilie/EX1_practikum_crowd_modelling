```python
import time
import numpy as np
import tensorflow as tf
from tensorflow.contrib.slim import fully_connected as fc
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)
num_sample = mnist.train.num_examples
input_dim = mnist.train.images[0].shape[0]
w = h = int(np.sqrt(input_dim))
```

```python
class VariantionalAutoencoder(object):

    def __init__(self, learning_rate=1e-3, batch_size=128, n_z=2):
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.n_z = n_z

        self.build()

        self.sess = tf.InteractiveSession()
        self.sess.run(tf.global_variables_initializer())

    def build(self):
        tf.reset_default_graph()
        self.x = tf.placeholder(
            name='x', dtype=tf.float32, shape=[None, input_dim])

        #2 fc with relu activation functions for encoding
        f1 = fc(self.x, 256, scope='enc_fc1', activation_fn=tf.nn.relu)
        f2 = fc(f1, 256, scope='enc_fc2', activation_fn=tf.nn.relu)
        
        #latent layer
        self.z_mu = fc(f2, self.n_z, scope='enc_fc4_mu', activation_fn=None)
        self.z_log_sigma_sq = fc(f2, self.n_z, scope='enc_fc4_sigma', activation_fn=None)
        eps = tf.random_normal(shape=tf.shape(self.z_log_sigma_sq), mean=0, stddev=1, dtype=tf.float32)
        self.z = self.z_mu + tf.sqrt(tf.exp(self.z_log_sigma_sq)) * eps
        
        #2 fc with activation function for decoding
        g1 = fc(self.z, 256, scope='dec_fc1', activation_fn=tf.nn.relu)
        g2 = fc(g1, 256, scope='dec_fc2', activation_fn=tf.nn.relu)
        self.x_hat = fc(g2, input_dim, scope='dec_fc4', activation_fn=tf.sigmoid)

        epsilon = 1e-10
        recon_loss = -tf.reduce_sum(self.x * tf.log(epsilon+self.x_hat) + (1-self.x) * tf.log(epsilon+1-self.x_hat), axis=1)
        self.recon_loss = tf.reduce_mean(recon_loss)

        latent_loss = -0.5 * tf.reduce_sum(1 + self.z_log_sigma_sq - tf.square(self.z_mu) - tf.exp(self.z_log_sigma_sq), axis=1)
        self.latent_loss = tf.reduce_mean(latent_loss)

        #ELBO
        self.total_loss = self.recon_loss + self.latent_loss
        self.train_op = tf.train.AdamOptimizer(learning_rate=self.learning_rate).minimize(self.total_loss)
        
        self.losses = {
            'recon_loss': self.recon_loss,
            'latent_loss': self.latent_loss,
            'total_loss': self.total_loss,
        }        
        return

    def run_single_step(self, x):
        _, losses = self.sess.run([self.train_op, self.losses], feed_dict={self.x: x})
        return losses

    # x -> x_hat
    def reconstructor(self, x):
        x_hat = self.sess.run(self.x_hat, feed_dict={self.x: x})
        return x_hat

    # z -> x
    def generator(self, z):
        x_hat = self.sess.run(self.x_hat, feed_dict={self.z: z})
        return x_hat
    
    # x -> z
    def transformer(self, x):
        z = self.sess.run(self.z, feed_dict={self.x: x})
        return z
```

```python
def trainer(model_class, learning_rate=1e-3, batch_size=128, num_epoch=5, n_z=2, log_step=1):
    model = model_class(learning_rate=learning_rate, batch_size=batch_size, n_z=n_z)

    elbo = []
    recon = []
    latent = []
    for epoch in range(num_epoch):
        start_time = time.time()
        
        for iter in range(num_sample // batch_size):
            batch = mnist.train.next_batch(batch_size)
            losses = model.run_single_step(batch[0])
            
        end_time = time.time()
        elbo.append(losses["total_loss"])
        recon.append(losses["recon_loss"])
        latent.append(losses["latent_loss"])
        
        if epoch % log_step == 0:
            log_str = '[Epoch {}] '.format(epoch)
            for k, v in losses.items():
                log_str += '{}: {:.3f}  '.format(k, v)
            log_str += '({:.3f} sec/epoch)'.format(end_time - start_time)
            print(log_str)
            
    print('Done!')

    plt.figure(figsize = (12, 12))
    plt.plot([e for e in range(num_epoch)], elbo, label = "ELBO")
    #plt.plot([e for e in range(num_epoch)], recon, label = "Reconstruction Loss")
    #plt.plot([e for e in range(num_epoch)], latent, label = "Latent Loss")
    plt.title("EPOCH VS ELBO")
    plt.legend()

    plt.show()
    
    return model
```

```python
def test_transformation(model_2d, mnist, batch_size=6000):
    
    assert model_2d.n_z == 2
    batch = mnist.test.next_batch(batch_size)
    z = model_2d.transformer(batch[0])
    plt.figure(figsize=(10, 8)) 
    plt.scatter(z[:, 0], z[:, 1], c=np.argmax(batch[1], 1), s=20)
    plt.colorbar()
    plt.grid()
```

```python
def test_reconstruction(model, mnist, h=28, w=28, batch_size=128):

    batch = mnist.test.next_batch(batch_size)
    x_reconstructed = model.reconstructor(batch[0])

    n = np.sqrt(batch_size).astype(np.int32)
    I_reconstructed = np.empty((h*n, 2*w*n))
    for i in range(n):
        for j in range(n):
            x = np.concatenate(
                (x_reconstructed[i*n+j, :].reshape(h, w), 
                 batch[0][i*n+j, :].reshape(h, w)),
                axis=1
            )
            I_reconstructed[i*h:(i+1)*h, j*2*w:(j+1)*2*w] = x

    plt.figure(figsize=(10, 20))
    plt.imshow(I_reconstructed, cmap='gray')
```

```python
def test_generation(model, z=None, h=28, w=28, batch_size=100):

    if z is None:
        z = np.random.normal(size=[batch_size, model.n_z])
    x_generated = model.generator(z)    

    n = np.sqrt(batch_size).astype(np.int32)
    I_generated = np.empty((h*n, w*n))
    for i in range(n):
        for j in range(n):
            I_generated[i*h:(i+1)*h, j*w:(j+1)*w] = x_generated[i*n+j, :].reshape(h, w)
            
    plt.figure(figsize=(8, 8))
    plt.imshow(I_generated, cmap='gray')
```

```python
print("dim=2 epoch=5")
model_vae_5 = trainer(VariantionalAutoencoder, n_z=2, num_epoch=5)
test_transformation(model_vae_5, mnist)
test_reconstruction(model_vae_5, mnist)
test_generation(model_vae_5)


```

```python
print("dim=2 epoch=25")
model_vae_25 = trainer(VariantionalAutoencoder, n_z=2, num_epoch=25)
test_transformation(model_vae_25, mnist)
test_reconstruction(model_vae_25, mnist)
test_generation(model_vae_25)
```
